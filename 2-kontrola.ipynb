Puna struktura grananja u pythonu je sljedeća:
```
if uvjet1:
   naredbe
elif uvjet2:
   naredbe
elif uvjet3:
   naredbe
else:
   naredbe
```

* if
   * započinje grananje
   * sadrži prvi uvjet
* elif
   * opcionalno
   * alternativna grana
   * testira se ako je prvi uvjet False
   * ima dodatni uvijet
* else
   * opcionalno
   * zadnja grana
   * ispunjava se ako je sve False
