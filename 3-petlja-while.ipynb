# Petlje

U prošlom zadatku smo igraču ponudili 3 pokušaja. No težina igre se može regulirati korištenjem većeg raspona
brojeva i većeg broja pokušaja. Veći raspon = teža, više pokušaja = lakša. No mi smo hardcodeali pokušaje,
pa to ne možemo mijenjati, a raspon možemo. Treba proširiti kako gledamo pokušaje - tu dolaze **petlje**.

Petlje su način na kojeg komad koda možemo ponavljati proizvoljni broj puta, bez da ponavljanje hardcodeamo.

Postoje dva tipa petlji:
* **for** x in y
    - Broj ponavljanja **jest** poznat prije ulaska u petlju
    - Može imati vlastiti brojač koraka
    - Dva načina postavljanja uvjeta:
        - **for element in niz**
            - Niz je bilo koji oblik kojeg python tretira kao uređeni niz.
            - List, set, dictionary - nismo radili
            - String (npr *for slovo in "python"* će vratiti jedno po jedno slovo)
        - **for broj in range(N)**
            - *range(N)* vraća niz brojeva od 0 do N-1 (ukupno N)
            - *broj* je ovdje može biti brojač
                - Ako nam je brojač ipak nepotreban, običaj je zamijeniti s _:
                    - **for _ in range(N)**
        - **for broj, element in enumerate(niz)**
            - Ako nam treba i brojač i elementi
* **while** uvijet
    - Broj ponavljanja **nije** poznat prije ulaska u petlju
    - Nema vlastiti brojač koraka
        - potrebno dodati kombinaciju
            - i = 0
            - i += 1
    - Izvršava se dok je uvijet točan
        - Ako *naš* uvijet označava kraj, moramo ga obrnuti
        - "stani ako health bude 0" = **while health > 0**
    - Može završiti kao **beskonačna petlja**

## Python blokovi

Većina programskih jezika odvaja zasebne blokove koristeći vitičaste zagrade {}. Python
odvaja koristeći uvlaku (indent) i dvotočku na početku bloka. To povećava preglednost koda,
ali je nezgodno za prvo učenje.

Prikaz koda koristeći uvlaku:

```python
broj_timova = int(input("Koliko pokušaja "))
# kraj vanjskog koda
for _ in range(broj_timova):
.    # početak prvog bloka
.    broj_igrača = int(input("Koliko igrača" ))
.    for __ in range(broj_igrača):
.    .    # početak drugog bloka
.    .    ime = input("Unesi ime igrača" )
.    .    dodaj_u_tim(ime)  # funkcija za dodavanje igrača
.    .    print("Unešeno ime igrača")
.    .    # kraj drugug bloka
.    print("Unesena imena igrača cijelog tima")
.    # kraj prvog bloka
# početak novog vanjskog koda
print("Unešeni su svi igrači i svi timovi")
```



## While petlja

Koristimo ju kad ne znamo točan broj ponavljanja, nego to ovisi o nekom uvjetu, "dok":
- Dok postoji input s tipkovnice
- Dok ima preostalih pokušaja
- Dok neprijatelj ima healtha
- Dok ne prekinem program

**Pažnja**: uvjet u while se koristi u pozitivnom smislu (kad bi petlja trebala raditi), a ne u negativnom (kad bi petlja trebala stati). Ako naša logika bolje opisuje negativnu ("stani kad mana padne na nula") moramo prebaciti na pozitivnu okretanjem logike ("nastavi dok je mana iznao nule").

Uvjet može biti bilo što što se može provjeriti, npr varijabla sama za sebe. Primjer: program koji pozdravlja sve dok se upisuje ime.
```python
ime = input("ime")
while ime:
    print("Dobrodošao,", ime)
    ime = input("ime")
```

Što ako izostavimo prije petlje? Što ako izostavimo input unutar petlje?

### While i brojač

While petlja nema brojač, pa se ili koristi za probleme gdje nije potrebno brojati, ili se brojač unosi samostalno. Dodajmo brojanje u prošli primjer:
```python
ime = input("ime")
i = 0
while ime:
    i += 1
    print("Dobrodošao,", ime)
    ime = input("ime")
print("predstavio si se", i, "puta")
```

## Beskonačna petlja

Specijalna situacija s while petljom je ako stavimo uvjet koji će uvijek biti True. Tada se petlja izvršava bez prekida, dok ne prekinemo sam program. Time je moguće petlju odvojiti na ponavljanje i uvjet. Za to nam treba i komanda za ručno prekidanje petlje: **break**.

Prepišimo gornji primjer na taj način:
```python
while True:
    ime = input("Unesi ime ")
    if ime:
        print("Dobrodošao,", ime)
    else:
        break
```

Razdvajanje ponavljanja i uvjeta je korisno ako imamo kompliciran uvjet, ili uvjet koji se kontrolira na više mjesta. Ali ne donosi ništa novo.
